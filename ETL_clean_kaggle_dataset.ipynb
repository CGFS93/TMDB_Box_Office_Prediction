```python
from ast import literal_eval
```

```python
movies_df = movies_df.dropna(subset=['genres'])
```

```python
def my_func(x):
    # if pd.isna(x):
        # return np.nan
    l=[]
    x=literal_eval(x)
    for i in x:
        l.append(i['name'])
    return l
movies_df['new_genre']=movies_df['genres'].apply(my_func)
```
